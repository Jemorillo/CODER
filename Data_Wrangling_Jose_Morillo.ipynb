```python
#Se transforma la columna 'geometry' en dos columnas 'latitud' y 'longitud' extrayendo el dato de ambos de un mismo campo:

df1['latitud'] = df['geometry'].apply(lambda x: x['location']['lat'])
df1['longitud'] = df['geometry'].apply(lambda x: x['location']['lng'])

#Se extrae de la columna 'type' el tipo de establecimiento: sea: bar, restaurante, nightclub, etc...

df1['type_st'] = df['types'].apply(lambda x: x[0] if len(x) > 0 else None)
df1.head()
```

```python
#Una vez extraidos los datos de las columnas,  se eliminan las columna 'geometry' y 'type'

df1.drop('geometry', axis=1, inplace=True)

df1.drop('types', axis=1, inplace=True)
df1.head()
```

```python
#Para la columna 'opening_hours' se requiere transformar a una nueva columna que señale si esta abierto o cerrado al momento de la extraccion de los datos mediante la API

# transformando los valores
def transf_estado(open_now):
    if open_now:
        return 'Open'
    else:
        return 'Closed'

# se crea una nueva columna 'estado' con los valores transformados
df1['estado'] = df['opening_hours'].apply(lambda x: transf_estado(x['open_now']))

#una vez obtenido el estado se elimina la columna 'opening_hours'
df1.drop('opening_hours', axis=1, inplace=True)
df1.head()
```

```python
#Para obtener una nueva columna que señale lalocalidad del negocio, se toma de la columna 'vicinity' el segundo elemento que contiene este dato requerido

#extraccion de direccion
df['address'] = df['vicinity'].str.split(', ').str[-2]

#extraccion de localidad
df['city'] = df['vicinity'].str.split(', ').str[-1]

df1.head()
```

```python
#se elimina la columna 'vicinity'
df1.drop('vicinity', axis=1, inplace=True)
```

```python
#se sustituyen los valores NaN de la columna 'price_level' por la media de los datos
df1['price_level'] = df1['price_level'].round(1)

media= df1['price_level'].mean()


# Reemplazar los valores NaN en la columna 'Edad' por la media
df1['price_level'].fillna(media, inplace=True)

df1.head()
```
